### Day of the week and Month analysis

```python
df['day']  = df['started_at'].dt.day_name()
c = df['day'].value_counts()
c
```

```python
plt.figure()
c.plot(kind = 'bar')
```

```python
pivot_by_day = df.pivot_table(index = 'day', columns='member_casual', values = 'ride_id', aggfunc=len, fill_value=0) # .pivot_table is used to make a pivot table from the columns and rows and ecah also categorizes on its own
pivot_by_day.plot(kind="bar")
```

On Saturdays and Sundays, casual users are more than members and the highest usage of services is on Saturdays

```python
df['started_at'].dt.month.value_counts()[:3]
```

This shows that July, August and September have the highest users

```python
df['month'] = df['started_at'].dt.month
k = df['month'].value_counts()
k[:3]
```

```python
pivot_by_month = df.pivot_table(index = 'month', columns='member_casual', values = 'ride_id', aggfunc=len, fill_value=0) # .pivot_table is used to make a pivot table from the columns and rows and ecah also categorizes on its own
pivot_by_month.plot(kind="bar")
```

In july and august, number of casual users outnumber the number of members

# Summary and Conclusion

- Overall, number of members are more than the casual users which means that there are more returning customers which is a positive thing.
- Month wise, July, August and September see the most active users and since that is mostly the vacation time and during that period the number of casual users exceed the number of members, pricing can slightly be increased for casual users during these months so as to increase marginal profit.
- Weekends see more rush than weekdays and the rush is mostly of casual users indicating that people use the services to go out during the weekends so weekend discounts can be given to members so as to convert casual users to members.
- Peak hours are from 4PM to 7PM most probably due to office rush and surge pricing can be applied for casual users while maintaing same price for members to again convert casual users to members.
- People utilising the services for more than 24 hours are mostly causual users and can be given discounts to encourage usage.
